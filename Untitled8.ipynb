```python
nested = [1, [2, [3, 4]], 5]

stack = nested[:]
flat = []

while stack:
    item = stack.pop()
    if isinstance(item, list):
        stack.extend(item)
    else:
        flat.append(item)

flat.reverse()
print(flat)
```

```python
s = "hello world python"

words = []
word = ""

for ch in s:
    if ch != " ":
        word += ch
    else:
        words.append(word)
        word = ""

words.append(word)

result = ""
for i in range(len(words)-1, -1, -1):
    result += words[i]
    if i != 0:
        result += " "

print(result)
```
